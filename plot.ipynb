```python
# Read csv

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
path = Path('build/training_data') / 'engagement_0_man0.csv'
df = pd.read_csv(path)
df.columns
```

```python
mx = df['m_x'].to_numpy()
my = df['m_y'].to_numpy()
tx = df['t_x'].to_numpy()
ty = df['t_y'].to_numpy()
plt.plot(mx, my)
plt.plot(tx, ty)
plt.grid()
```

```python
a = df['t_a'].to_numpy()
time = df['time'].to_numpy()
time
```

```python
# Get closest miss distance
rel_r = df['rel_r'].to_numpy()
min_miss_distance = rel_r.min()
min_idx = rel_r.argmin()
time_at_min = df['time'].iloc[min_idx]

print(f"Closest Miss Distance: {min_miss_distance:.4f} meters")
print(f"Occurred at time: {time_at_min:.4f} seconds")
print(f"Final range: {rel_r[-1]:.4f} meters")
# print(f"Simulation duration: {df['time'].iloc[-1]})
```

```python
# Missile 
m_vx = df['m_vx'].to_numpy()
m_vy  = df['m_vy'].to_numpy()

m_v = np.linalg.norm([m_vx, m_vy], axis=0)
m_v
```
