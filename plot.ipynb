```python
# Read csv

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
path = Path('build') / 'sim.csv'
df = pd.read_csv(path)
df.columns
```

```python
mx = df['m_x'].to_numpy()
my = df['m_y'].to_numpy()
tx = df['t_x'].to_numpy()
ty = df['t_y'].to_numpy()
```

```python
plt.plot(mx, my)
plt.plot(tx, ty)
plt.grid()
```

```python
xh_lam = df['xh_xlam'].to_numpy()
xlam = df['rel_xlam'].to_numpy()
time = df['time'].to_numpy()

plt.plot(time, xlam)
plt.plot(time, xh_lam)
plt.grid()
```
